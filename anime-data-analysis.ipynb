```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"anime.csv")
```

```python
# make a new column for episode count
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")":
            return data
        if check == True:
            data += i
        if i == "(":
            check = True
```

```python
df["Episodes"] = df["Title"].apply(extract_episodes)
df["Episodes"] = df["Episodes"].str.replace(" eps","")
df["Episodes"] = df["Episodes"].astype(int)
```

```python
df
```

```python
# make a new column for time stamp
```

```python
df.loc[0]["Title"]
```

```python
def extract_time(txt):
    check = False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ")":
            for j in range(i+1,i+20):
                data += txt[j]
            return data
```

```python
df["Time stamp"] = df["Title"].apply(extract_time)
```

```python
df.head()
```

```python
# Now extract Months from the Years
```

```python
# Used AI in this.......
def calculate_months(time_str):
    if pd.isna(time_str):
        return None
    parts = str(time_str).split(" - ")
    if len(parts) != 2:
        return None
    start = pd.to_datetime(parts[0], format="%b %Y", errors="coerce")
    end = pd.to_datetime(parts[1], format="%b %Y", errors="coerce")
    if pd.isna(start) or pd.isna(end):
        return None
    return (end.year - start.year) * 12 + (end.month - start.month) + 1
    
```

```python
df["Months"] = df["Time stamp"].apply(calculate_months)
```
